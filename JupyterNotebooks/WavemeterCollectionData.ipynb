```python
import pandas as pd
import os 
import sys 
from openpyxl import load_workbook

work_dir =  'N:/Experiments/InstrumentControl/'
path = os.path.abspath(work_dir)
if not path in sys.path:
    sys.path.insert(0, path)
    print(path)
    
from pyWavemeter import Bristol 
```

```python
# -- initialize the file --
basepath = 'N:/Experiments/Users/Greg/Projects/PhotonicsFlywheel/2021-05-24-LinearMeasurements/'
xlsname = 'LInearWavemeter_newLaser.xlsx'
df = pd.DataFrame(columns = ['fres', 'µ'],)
writer = pd.ExcelWriter(f'{basepath}{xlsname}', engine='xlsxwriter')
df.to_excel(writer, sheet_name='Sheet1', index=False)
writer.save()
```

# ONLY RUN BELOW After init

```python
# -- Append to the Excel file

book = load_workbook(f'{basepath}{xlsname}')
writer = pd.ExcelWriter(f'{basepath}{xlsname}', engine='openpyxl')
writer.book = book
writer.sheets = {ws.title: ws for ws in book.worksheets}

with Bristol(ip = '10.0.0.29') as wvmtr: 
    freq = wvmtr.freq
print(freq)

df1 = pd.DataFrame(dict(fres = [freq]), index = [0])
for sheetname in writer.sheets:
    df1.to_excel(writer,sheet_name=sheetname, startrow=writer.sheets[sheetname].max_row, index = False,header= False)

writer.save()

df = pd.read_excel(f'{basepath}{xlsname}')
df
```
